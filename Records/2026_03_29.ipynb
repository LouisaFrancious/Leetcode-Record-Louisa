## 15. 3 Sum
Similar to 2 sum

### 那我能不能用2sum的逻辑来做3sum？

用left和right 只要left<right 就一直尝试




[1,2,0,1,0,0,0,0]

### 首先可以用sort()

然后用i固定一个数，来确定一个位置先：

    那么后面的问题就可以变成找两个数，让他们加起来等于-nums[i]

一定要用排序，因为排序的话就知道怎么移动了

for i in range(0,len(nums)-2):

### 定左和右:

left = i+1, right = len(nums)-1

用while loop来循环left and right :如果没有while loop if条件只会循环一次

根据total = nums[left]+nums[right]+nums[i]来判断

如果total>0 说明总和太大了 需要变小 left往右移动数才会变大

while left< right保证了固定i之后 把i后面所有可能的left和right都试完

for 负责举第一个数

while 负责在后面不断尝试两个数

### 为什么外层只循环到len(nums)-2
因为我要找三个数

如果i已经走到倒数第二个数了 后面只剩一个数字 right就不够了

做题的时候要注意 写循环范围的时候 不能机械的写len(nums) 要想清楚后面还有几个元素

### 去重
有两种情况: 如果重复两个出现 那么跳过

不要等结果重复了再处理 再生成重复之前就跳过



```python
class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        nums.sort() #排序
        answer = [] #记录下来answer

        for i in range(0,len(nums)-2):
            if i>0 and nums[i]==nums[i-1]:
                continue

            left = i+1
            right = len(nums)-1

            while left<right:#在for循环中，left和right分别指向i的下一个元素和最后一个元素，进行双指针遍历
                total =nums[i]+nums[left]+nums[right]

                if total==0:
                    answer.append([nums[i],nums[left],nums[right]]) #记录答案
                    left+=1 #移动left指针
                    right-=1 #移动right指针

                    while left < right and nums[left] == nums[left - 1]:
                        left += 1 #跳过重复的元素

                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1 #跳过重复的元素

                elif total>0:
                    right-=1
                    
                elif total<0:
                    left+=1


        return answer
```
